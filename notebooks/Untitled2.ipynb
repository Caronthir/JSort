# Alignment

```julia
using Revise
push!(LOAD_PATH, "$(homedir())/Development/")
using JSort
```

```julia
path = "/home/erdos/gits/JSort/si/"
j(x) = joinpath(path, x)
det = ChannelDetector("eg", path);
```

```julia
plot(det)
```

```julia
lowregion=135:1300
highregion=1450:2600
al, calib1, icalib1 = align(det, lowregion=lowregion, highregion=highregion, lownumregions=5,
           lowsearchwidth=80, highsearchwidth=150, highregionwidth=200);
plot(al)
```

```julia
al2, calib2, icalib2 = align(al, lowregion=100:1500, highregion=1400:2500,
                   lowsearchwidth=70, lowregionwidth=150,
                   highsearchwidth=80, highregionwidth=200, lownumregions=10, 
                    highnumregions=5)
plot(al2)
```

```julia
write(j("coefficients_g_1.csv"), calib1)
write(j("coefficients_g_inv_1.csv"), icalib1)
write(j("coefficients_g_2.csv"), calib2)
write(j("coefficients_g_inv_2.csv"), icalib2)
```
